```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt=sns.countplot(x=ap['clusters'])
plt.set_title("Jumlah Kota Tiap Kluster")
plt.spines[['right', 'top']].set_visible(False)
```

```python
ap1=ap[ap['clusters']=='1']
ap1
```

```python
ap1.describe()
```

```python
x = ap1[['SO2', 'temp', 'wind', 'precip','predays']].corr()
plt = sns.heatmap(x, annot=True)
plt.set_title("Korelasi antar variabel")
```

```python
plt.plot(ap1['City'], ap1['temp'], label='temp')
plt.plot(ap1['City'], ap1['precip'], label='precip')
plt.title('temperatur & curah hujan')
plt.xlabel('City')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Pada kluster 1 terlihat bahwa pada grafik yang temparatur (biru) dan precipitation (oren) keduanya tidak menunjukkan perbedaan yang jauh antar satu kota dengan kota lainnya. Rata-rata curah hujan tahunan (precip) tiap kota di kluster ini berkisar antara 47 sampai 56 inch yang mengindikasikan curah hujan di kota-kota pada kluster ini cukup tinggi karena lebih dari rata-rata curah hujan tahunan USA (30.21 inch). Selain itu rata-rata suhu tahunan di kluster ini berkisar antara 47 sampai 57 F yang mengindikasikan kota-kota di kluster ini memiliki suhu normal jika dibandingkan dengan rata-rata suhu di USA (52.7 F).
